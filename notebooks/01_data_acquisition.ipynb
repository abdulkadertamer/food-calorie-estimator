## 🎯 Step 6 — Target Variable: Calories

**الـ Target = calories**  
العلاقة الرياضية: `Calories = (protein × 4) + (carbs × 4) + (fat × 9)`

```python
print('🎯 Target Variable: calories')
print(f'Type  : {df["calories"].dtype}')
print(f'Min   : {df["calories"].min()}')
print(f'Max   : {df["calories"].max()}')
print(f'Mean  : {df["calories"].mean():.2f}')
print(f'Nulls : {df["calories"].isnull().sum()}')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].hist(df['calories'].dropna(), bins=50, color='#FF6B6B', edgecolor='white', alpha=0.8)
axes[0].set_title('📊 Distribution of Calories', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Calories')
axes[0].set_ylabel('Count')

axes[1].boxplot(df['calories'].dropna(), patch_artist=True,
                boxprops=dict(facecolor='#4ECDC4', color='#2C3E50'),
                medianprops=dict(color='#E74C3C', linewidth=2))
axes[1].set_title('📦 Boxplot of Calories', fontsize=14, fontweight='bold')
axes[1].set_ylabel('Calories')

plt.tight_layout()
os.makedirs('../results/plots', exist_ok=True)
plt.savefig('../results/plots/01_calories_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
print('✅ Plot saved!')
```

## 📊 Step 7 — Missing Values

```python
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
missing_df = pd.DataFrame({'Missing Count': missing, 'Missing %': missing_pct})
missing_df = missing_df[missing_df['Missing Count'] > 0].sort_values('Missing %', ascending=False)

print(f'⚠️  Columns with missing values: {len(missing_df)}')
print(f'✅ Clean columns: {len(df.columns) - len(missing_df)}')
print(missing_df.head(15).to_string())
```
